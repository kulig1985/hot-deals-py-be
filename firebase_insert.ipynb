```python
result_df_list = []

for file in files:
    if file.endswith('.xlsx'):
        print(f'file: {file}')
        df = pd.read_excel(file)
        result_df_list.append(df)
        

df_all_result = reduce(lambda left, right: pd.concat([left, right]), result_df_list)
df_all_result = df_all_result.reset_index()
del df_all_result['index']

df_all_result['itemName'] = df_all_result['itemName'].str.replace('\n', '')
df_all_result['itemCleanName'] = df_all_result['itemCleanName'].str.replace('\n', '')
df_all_result['measure'] = df_all_result['measure'].fillna('N.a')
```
