```python
def get_all_tesco_link():
    
    fake_headers = {'Host': 'bevasarlas.tesco.hu',
               'User-Agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:102.0) Gecko/20100101 Firefox/102.0',
               'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
               'Accept-Language': 'hu-HU,hu;q=0.8,en-US;q=0.5,en;q=0.3',
               'Accept-Encoding': 'gzip, deflate, br',
               'Connection': 'keep-alive',
               'Referer': 'https://www.spar.hu/',
               'Sec-Fetch-Dest': 'empty',
               'Sec-Fetch-Mode': 'cors',
               'Sec-Fetch-Site': 'none',
               'Upgrade-Insecure-Requests': '1',
               'Sec-Fetch-User': '?1',
               'TE': 'trailers'}
    offer_page = 'https://bevasarlas.tesco.hu/groceries/hu-HU/'
    
    r = requests.get(offer_page,headers=fake_headers)
    soup = bs(r.content)
    
    divs_body = soup.body.findAll('div', {'class' : ['menu-tree main-level-menu-tree--open menu-tree--hide-down-chevrons']})
    
    all_link = []
    for div in divs_body:
        for a in div.find_all('a', href=True):
            find_value = a['href']
            find_value = find_value[:find_value.find('?')]
            url = 'https://bevasarlas.tesco.hu' + find_value + '/all?include-children=true'
            #print(url)
            all_link.append(url)
            
            
    link_extend= set()
    
    for link in all_link:
        print(link)
        r = requests.get(link,headers=fake_headers)
        soup = bs(r.content)

        divs_pagi = soup.body.findAll('div', {'class' : ['pagination__items-displayed']})

        for div in divs_pagi:
            for strong in div.findAll('strong'):
                if strong.text.find('termékből') > -1:
                    number = int(re.findall(r'\d+', strong.text)[0])
                    const = 20
                    div = number // const
                    div_rem = number % const
                    query = '&page={num}&count={const}'
                    for x in range(1,div+2):
                        url_extend = link + query.format(num=x,const=const)
                        print(url_extend)
                        link_extend.add(url_extend)
    
    return list(link_extend)
```

```python
tesco_links = get_all_tesco_link()
```

```python
def find_nth_occurrence(string, char, occurrence):

    val = -1
    for i in range(0, occurrence):
        val = string.find(char, val + 1)
    return val
```

```python
def image_download(img_url):
    print('invoked..')
    uuid_image = uuid.uuid1()
    res = requests.get(img_url)

    if res.headers['Content-Type'][res.headers['Content-Type'].find('/')+1:] == 'jpeg':

        image_name = str(uuid_image) + '.jpg'

    if res.headers['Content-Type'][res.headers['Content-Type'].find('/')+1:] == 'plain':

        image_name = str(uuid_image) + '.png'
    
    opener = urllib.request.build_opener()
    opener.addheaders = [('Host', 'cdn1.interspar.at'), ('User-Agent' , 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:103.0) Gecko/20100101 Firefox/103.0')]
    urllib.request.install_opener(opener)

    img_path = '/Users/kuligabor/Documents/HotDealsHungary/img/' + image_name
    urllib.request.urlretrieve(img_url , img_path)
    
    return image_name
```

```python
all_items = []
counter = 1

for t_link in tesco_links:
    
    print(f'crawl url: {t_link} done {counter} from {len(tesco_links)}')
    counter = counter + 1
    
    r = requests.get(t_link,headers=fake_headers)
    soup = bs(r.content)
    
    divs_list_page = soup.body.findAll('div', {'class' : ['product-lists-wrapper']})
    
    for div in divs_list_page:
        
        cards = div.findAll('li', {'class' : 'product-list--list-item'})
        for card in cards:
            try:
                item_dict = {}

                id_link = card.find('a', {'class' : 'product-image-wrapper'})['href']
                item_dict['itemId'] = id_link[find_nth_occurrence(id_link,'/',4)+1:]
                name = card.find('span', {'class' : 'beans-link__text'})
                item_dict['itemName'] = name.text

                item_dict['itemCleanName'] = unidecode.unidecode(item_dict['itemName']).lower()

                img = card.find('img', {'class' : 'product-image'})
                img_url = img['srcset']
                #item_dict['imageUrl'] = img_url[:img_url.find(' ')] 
                item_dict['imageUrl'] = image_download(img_url[:img_url.find(' ')])

                price = card.find('p', {'class' : 'beans-price__text'}) 
                item_dict['price'] = price.text.replace(' Ft', '').replace(' ','').strip()

                measure = card.find('p', {'class' : 'beans-price__subtext'}) 
                item_dict['measure'] = measure.text

                item_dict['salesStart'] = np.nan

                item_dict['source'] = t_link[t_link.find('/shop'):find_nth_occurrence(t_link, '/', 7)].replace('/shop/', '')

                item_dict['runDate'] = datetime.now().strftime('%Y.%m.%d-%H:%M:%S')
                item_dict['shopName'] = 'tesco'

                special = card.find('div', {'class' : 'special-offer-sash'})

                if special != None:
                    item_dict['isSales'] = 1
                else:
                    item_dict['isSales'] = 0

                all_items.append(item_dict)
                
            except Exception as e:
                print(e)
                continue 
            
df = pd.DataFrame(all_items)
```
